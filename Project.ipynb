```python
def clean_age(age):
    try:
        return float(age)
    except ValueError:
        return float('nan')

df['Age'] = df['Age'].apply(clean_age)
```

```python
under_30_filter = df['Age'] < 30
```

```python
df_under_30 = df.loc[under_30_filter]
```

```python
df[under_30_filter]
```

```python
combined_filter = under_30_filter & country_filter
```

```python
df = df[combined_filter]
```

```python
import numpy as np
import matplotlib.pyplot as plt

ages = df['Age'].values
compensations = df['ConvertedComp'].values

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.bar(np.arange(len(ages)), ages, label='Age')
ax1.set_xlabel('Respondent')
ax1.set_ylabel('Age')
ax1.set_title('Age Distribution')
ax1.legend()


plt.show()
```
